```python
# Importing...
import numpy as np
import pylab as pl
import sys as sys

data = np.loadtxt("C:/Users/Blake/Downloads/cpresources/circular.txt",float)


#plots the data
pl.imshow(data)
pl.show()


#Creates a 2d Fourier transform and images it.
ftimage = np.fft.fft2(data)

#This centers the transformed image
ftimage = np.fft.fftshift(ftimage)

pl.imshow(np.abs(ftimage))
pl.show()


#Creates an inverse transform based on the Forier transform above and images it.
iimage = np.fft.ifft2(ftimage)
pl.imshow(np.abs(iimage))
pl.show()

```

```python
# Importing numpy and matplotlib.pyplot, and the relevent scatter plot stuff form pylab
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sc

# I downloaded the data from http://www-personal.umich.edu/~mejn/cp/programs.html
# I selected the stars.txt file because I like space stuff.
# This took forever to figure out.
# When using the import loadtxt function you must use either forward slashes in the directory 
# or double slashes.
data = np.loadtxt("C:/Users/Blake/Downloads/cpresources/stars.txt",float)

# Since the data is in two columns, each needed to be defined as either an x or y value.
x = data[:,0]
y = data[:,1]

#define the fitting function
f = lambda x,a,c: a*(-x)+c 

#define the original fit parameters
x_min, x_max = 0, 12000
a,c = 1/2, 15

#Compute the fit
params, extras = sc.curve_fit(f, x, y)


# Creates the scatterplot
plt.scatter(x,y)
plt.plot(x,f(x,params[0],params[1]))

#lables axes
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.xlim(0,13000)
plt.ylim(-5,20)
plt.show()

# Displays fit paramters
print('original coefficients: %6.3f, %6.3f' %(a,c))
print('fitted   coefficients: %6.3f, %6.3f' %(params[0], params[1]))
```

```python
#Part 1

import matplotlib.pyplot as plt
import numpy as np
from random import random

Graph = plt.figure()
ax1 = Graph.add_subplot(111, aspect='equal')

x1 = np.linspace(0,2,100)

#Circle R=1
y1 = np.sqrt(1-(x1-1)**2)
ax1.plot(x1,y1,'k')

#Circle R=2
y2 = 2-np.sqrt(4-x1**2)
ax1.plot(x1,y2,'k')

ax1.set_xlim(0,2)
ax1.set_ylim(0,2)

ax1.fill_between(x1, y1, y2, where=y1>y2, facecolor='green')

#Creates a boundary around the desired area
plt.plot([0,0,1.6,1.6,0],[0,1,1,0,0])
plt.show()

#Defines the boundaries of the figure in the box.
def Bottom(x):
    return 2-np.sqrt(4-x**2)
def Top(x):
     return np.sqrt(1-(x-1)**2)

#Performs the MC I.
def Prob_count(N):
    count = 0
    for i in range (N):
            x=2*random()
            y=2*random()
            if Bottom(x)<y and y<Top(x):
                count += 1
    I= 4*count/N
    return I

print("For 100 values: ", Prob_count(100))
print("For 1000 values: ", Prob_count(1000))
print("For 10000 values: ", Prob_count(10000))
print("For 100000 values: ", Prob_count(100000))
print("For 1000000 values: ", Prob_count(1000000))

```

```python
#From 13.4

import matplotlib.pyplot as plt
import numpy as np
from math import sin
from random import random

def f(x):
    return np.sin(1/x/(2-x))**2

# plot the f(x)
x = np.linspace(0.001,2,100)
plt.plot(x, f(x))
# plot the boundary
plt.plot([0,0,2,2,0],[0,1,1,0,0])
plt.show()

# MC integral
def f(x):
    return sin(1/x/(2-x))**2
def MC(N):
    count = 0
    for i in range(N):
        x = 2*random()
        y = random()
        if y<f(x):
            count += 1
    I = 2*count/N
    return I

#Creates and stores data in lists
MC100 = []
MC1000 = []
MC10000 = []
MC100000 = []
MC1000000= []

for z in range(100):
    MC100.append(MC(100))
    MC1000.append(MC(1000))
    MC10000.append(MC(10000))
    MC100000.append(MC(100000))
    MC1000000.append(MC(1000000))
    
MC100 = np.array(MC100)

#prints it all  
print("For 100 values: ", MC(100), "Standard Variance: ", np.sqrt(np.std(MC100)))
print("For 1000 values: ", MC(1000), "Standard Variance: ", np.sqrt(np.std(MC1000)))
print("For 10000 values: ", MC(10000), "Standard Variance: ", np.sqrt(np.std(MC10000)))
print("For 100000 values: ", MC(100000), "Standard Variance: ", np.sqrt(np.std(MC100000)))
print("For 1000000 values: ", MC(1000000), "Standard Variance: ", np.sqrt(np.std(MC1000000)))
```
